```python
# Create new column age_group to divide ages into differnt groups
```

```python
def age_group(x):
    if x >= 60:
        return "Senior"
    if x >= 40:
        return "Middle-aged"
    if x>= 30:
        return "Adult"
    if x<30:
        return "Young Adult"
        
```

```python
df['age_group']=df['age'].apply(age_group)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
#convert frequency_of_purchases into interger values for better analysis
```

```python
func_mapping = {
    "Fortnightly" :14,
    "Weekly" : 7,
    "Annually": 365,
    "Quarterly": 90,
    "Bi-Weekly": 14,
    "Monthly":30,
    "Every 3 Months":90



}
```

```python
df['purchase_frequency_days']=df['frequency_of_purchases'].map(func_mapping)
```

```python
df['purchase_frequency_days'].unique()
```
